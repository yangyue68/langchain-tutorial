1、使用构造函数创建一个提示模板

```python
from langchain_core.prompts import PromptTemplate

template=PromptTemplate(input_variables=["adjective","name"],template="你是一个{adjective}的人，你的名字叫： {name}")
print(template.format(adjective="good",name="zhangsan"))
print(template)
```

2、使用from_template方法创建一个提示模板

```python
from langchain_core.prompts import PromptTemplate

template=PromptTemplate.from_template("你是一个{adjective}的人，你的名字叫： {name}")
print(template.format(adjective="good",name="zhangsan"))
```

3、实例化部分提示词

```python
from langchain_core.prompts import PromptTemplate

template=PromptTemplate(input_variables=["adjective","name"],template="你是一个{adjective}的人，你的名字叫： {name}",
                        partial_variables={"adjective":"good"})
print(template.format(name="zhangsan"))
```

4、使用partial（） 方法在已有的模板上创建模板

```python
from langchain_core.prompts import PromptTemplate

template=PromptTemplate.from_template("你是一个{adjective}的人，你的名字叫： {name}")
template2=template.partial(adjective="good")
print(template2.format(name="zhangsan"))
```

5、组合提示词

```python
from langchain_core.prompts import PromptTemplate
template1=(PromptTemplate.from_template("你是一个{adjective}的人，你的名字叫： {name}")
           +"\n请给我将一个笑话"
           +"\n使用{language}语言")
print(template1.format(adjective="good",name="zhangsan",language="中文"))
```

6、invoke方法，返回的是一个PromptValue类型，要是有to_string()方法转换成字符串

```python
from langchain_core.prompts import PromptTemplate
template=PromptTemplate.from_template("你是一个{adjective}的人，你的名字叫： {name}")
print(template.invoke({"adjective":"good","name":"zhangsan"}))
print(template.invoke({"adjective":"good","name":"zhangsan"}).to_string())
```

7、结合大模型使用

```python
from langchain_core.prompts import PromptTemplate
from langchain_openai import OpenAI
import dotenv
import os
dotenv.load_dotenv()
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
os.environ["OPENAI_BASE_URL"]=os.getenv("OPENAI_BASE_URL")

model=OpenAI()
template=PromptTemplate.from_template("请评价一下{product}的优点和缺点，包括{aspect1}和{aspect1}两个方面，不要只说优点")
result=model.invoke(template.invoke({"product":"iphone 14 pro","aspect1":"价格","aspect2":"性能"}))
print(result)
```
