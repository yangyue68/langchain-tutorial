4、调用模板的四种方法：
1.invoke方法
2.format方法
3.format_messages方法
4.format_prompt方法

4.1 invoke方法
输入是字典，输出是ChatPromptValue

```python
from langchain_core.prompts import ChatPromptTemplate
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant.your name is {name}"),
    ("human", "What is the meaning of life?"),
    ("ai", "The meaning of life is to live."),
    ("human", "thank you,{user_input}?"),
])
prompt=chat_template.invoke(input={"name": "Alice","user_input":"你为什么这么帅？"})
print(prompt.to_messages())
print(prompt)
print(type(prompt))#<class 'langchain_core.prompt_values.ChatPromptValue'>
```

4.2 format方法
输入是变量的值
输出是字符串，字符串没有to_messages方法

```python
from langchain_core.prompts import ChatPromptTemplate
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant.your name is {name}"),
    ("human", "What is the meaning of life?"),
    ("ai", "The meaning of life is to live."),
    ("human", "thank you,{user_input}?"),
])
prompt=chat_template.format(name="Alice",user_input="你为什么这么帅？")
# print(prompt.to_messages())  #AttributeError: 'str' object has no attribute 'to_messages'
print(prompt)
print(type(prompt))#<class 'str'>
```

4.3 format_messages方法
输入是变量的值
输出是列表，列表没有to_messages方法，列表元素是SystemMessage、HumanMessage、AIMessage等

```python
from langchain_core.prompts import ChatPromptTemplate
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant.your name is {name}"),
    ("human", "What is the meaning of life?"),
    ("ai", "The meaning of life is to live."),
    ("human", "thank you,{user_input}?"),
])
prompt=chat_template.format_messages(name="Alice",user_input="你为什么这么帅？")
# print(prompt.to_messages())#AttributeError: 'list' object has no attribute 'to_messages'
print(prompt)
print(type(prompt))#<class 'list'>
```

4.4 format_template方法
输入是变量的值
输出是ChatPromptValue,可以调用to_messages方法

```python
from langchain_core.prompts import ChatPromptTemplate
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant.your name is {name}"),
    ("human", "What is the meaning of life?"),
    ("ai", "The meaning of life is to live."),
    ("human", "thank you,{user_input}?"),
])
prompt=chat_template.format_prompt(name="Alice",user_input="你为什么这么帅？")
print(prompt.to_messages())
print(prompt)
print(type(prompt))#<class 'langchain_core.prompt_values.ChatPromptValue'>
```

5、更丰富的参数类型：
不论是使用构造函数还是使用from_messages方法创建提示词模板，都可以使用更丰富的参数类型。
参数都是列表形式，但是列表元素类型非常丰富，包括：字符串、字典、字符串构成的元组、消息类型、提示词模板类型、消息提示词模板类型

5.1str类型，不推荐，因为角色都默认是human

```python
from langchain_core.prompts import ChatPromptTemplate
chat_template = ChatPromptTemplate.from_messages([
    "hello,{name}"#等价于：("human","hello,{name}")
])
prompt=chat_template.invoke(input={"name": "Alice"})
print(prompt.to_messages())
print(prompt)
print(type(prompt))
```

5.2字典类型

```python
from langchain_core.prompts import  ChatPromptTemplate
chat_template = ChatPromptTemplate.from_messages([
    {"role": "system", "content": "You are a helpful assistant.your name is {name}"},
    {"role": "human", "content": "What is the meaning of life?"},
    {"role": "ai", "content": "The meaning of life is to live."},
])
prompt=chat_template.invoke(input={"name": "Alice"})
print(prompt.to_messages())
print(prompt)
print(type(prompt))
```

5.3Message类型。
在指定角色时不能使用Message类型，只能使用字符串类型.
但是可以不指定角色，直接使用Message类型,这时角色显然是已知的

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.messages import SystemMessage,HumanMessage,AIMessage
chat_template = ChatPromptTemplate.from_messages([
    SystemMessage(content="You are a helpful assistant.your name is {name}"),
    HumanMessage(content="What is the meaning of life?"),
    AIMessage(content="The meaning of life is to live.")
])
prompt=chat_template.invoke(input={"name": "Alice"})
print(prompt.to_messages())
print(prompt)
print(type(prompt))
```

5.4ChatPromptTemplate类型里面嵌套ChatPromptTemplate类型

```python
from langchain_core.prompts import ChatPromptTemplate
nested_template1=ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant.your name is {name}"),
])
nested_template2=ChatPromptTemplate.from_messages([
    ("human", "What is the meaning of life?")
])
nested_template3=ChatPromptTemplate.from_messages([
    ("ai", "The meaning of life is to live.")
])
chat_template = ChatPromptTemplate.from_messages([
    nested_template1,
    nested_template2,
    nested_template3
])
prompt=chat_template.invoke(input={"name": "Alice"})
print(prompt.to_messages())
print(prompt)
print(type(prompt))
```

5.5ChatPromptTemplate类型里面嵌套SystemMessagePromptTemplate、HumanMessagePromptTemplate、AIMessagePromptTemplate

```python
from langchain_core.prompts import ChatPromptTemplate,SystemMessagePromptTemplate,HumanMessagePromptTemplate,AIMessagePromptTemplate

nested_template1=SystemMessagePromptTemplate.from_template("You are a helpful assistant.your name is {name}")
nested_template2=HumanMessagePromptTemplate.from_template("What is the meaning of life?")
nested_template3=AIMessagePromptTemplate.from_template("The meaning of life is to live.")
chat_template = ChatPromptTemplate.from_messages([
    nested_template1,
    nested_template2,
    nested_template3
])
prompt=chat_template.invoke(input={"name": "Alice"})
print(prompt.to_messages())
print(prompt)
print(type(prompt))
```

6、综合使用

```python
from langchain_core.messages import SystemMessage,HumanMessage,AIMessage
from langchain_core.prompts import ChatPromptTemplate,SystemMessagePromptTemplate,HumanMessagePromptTemplate,AIMessagePromptTemplate

system_prompt=SystemMessagePromptTemplate.from_template("You are a helpful assistant.your name is {name}")
human_prompt=HumanMessagePromptTemplate.from_template("What is the meaning of life?")
ai_prompt=AIMessagePromptTemplate.from_template("The meaning of life is to live.")

system_msg=SystemMessage(content="你是不是很迷茫？")
human_msg=HumanMessage(content="是的，我需要帮助")
ai_msg=AIMessage(content="那我帮你找一个")

nested_template=ChatPromptTemplate.from_messages([
    ("system","你对哪些方面感到迷茫？")
])

prompt_template=ChatPromptTemplate.from_messages([
    system_prompt,
    human_prompt,
    ai_prompt,
    system_msg,
    human_msg,
    ai_msg,
    nested_template
])
prompt=prompt_template.invoke(input={"name": "Alice"})
print(prompt.to_messages())
print("\n\n")
print(prompt)
print("\n\n")
print(type(prompt))
```
