```python
from langchain_openai import ChatOpenAI
from langchain.agents import AgentType,initialize_agent
from langchain.tools import Tool
from langchain_community.tools.tavily_search import TavilySearchResults
from langchain_experimental.utilities.python import PythonREPL
import os
import dotenv
dotenv.load_dotenv()
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["TAVILY_API_KEY"] = os.getenv("TAVILY_API_KEY")
os.environ["OPENAI_BASE_URL"] = os.getenv("OPENAI_BASE_URL")

#搜索工具
search=TavilySearchResults(max_results=3)
search_tool=Tool(
    name="Search",
    description="Search the web for information and use the result to answer the user's question.",
    func=search.run
)

#计算工具
python_repl=PythonREPL()
calc_tool=Tool(
    name="Calculator",
    description="A calculator. Useful for when you need to answer questions about math. Input should be a valid math question that can be translated into a valid python expression.",
    func=python_repl.run
)

agent_executor=initialize_agent(
    tools=[search_tool,calc_tool],
    llm=ChatOpenAI(temperature=0,model="gpt-4o-mini"),
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

results=agent_executor.run("特斯拉今年股价是多少？比去年上涨了百分之几？")
print(results)
```
