```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain.chains import LLMChain
import dotenv
import os
dotenv.load_dotenv()
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["OPENAI_BASE_URL"] = os.getenv("OPENAI_BASE_URL")
model = ChatOpenAI(model="gpt-4o-mini", temperature=0.9)
prompt = ChatPromptTemplate.from_messages([("system","你是一个牛逼的数学家"),("human","请计算{formulation}")])
llm_chain=LLMChain(llm=model,prompt=prompt)
result=llm_chain.invoke({"formulation":"1+1"})
print(result)
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain.chains import LLMChain
import dotenv
import os
dotenv.load_dotenv()
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["OPENAI_BASE_URL"] = os.getenv("OPENAI_BASE_URL")
model = ChatOpenAI(model="gpt-4o-mini", temperature=0.9)
prompt = ChatPromptTemplate.from_messages([("system","你是一个牛逼的计算机工程师"),("human","请计算关于{topic}的笑话")])
llm_chain=LLMChain(llm=model,prompt=prompt,verbose=True)
result=llm_chain.invoke({"topic":"结婚"})
print(result)
```
