```python


from langchain_openai import ChatOpenAI
from langchain.chains import SimpleSequentialChain,LLMChain
from langchain.prompts import ChatPromptTemplate
import os
import dotenv
dotenv.load_dotenv()
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["OPENAI_BASE_URL"] = os.getenv("OPENAI_BASE_URL")
model = ChatOpenAI(model="gpt-4o-mini", temperature=0)

template1="""
你是一个剧作家，给定剧本的标题，你的工作就是根据这个标题写一个大纲。title:{title}
"""
prompt1 = ChatPromptTemplate.from_template(template1)
chain1 = LLMChain(llm=model, prompt=prompt1)

template2="""
你是一个剧评家，给定一个大纲，你需要根据大纲来写一篇评论。 outline:{outline}
"""
prompt2 = ChatPromptTemplate.from_template(template2)
chain2 = LLMChain(llm=model, prompt=prompt2)
full_chain=SimpleSequentialChain(chains=[chain1, chain2], verbose=True)
result=full_chain.invoke({"input":"《禁忌的爱》"})
print(result)
```
